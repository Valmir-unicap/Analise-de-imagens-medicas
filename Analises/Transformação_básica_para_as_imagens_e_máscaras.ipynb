```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms

def mostrar_imagem(imagem, mascara, titulo_imagem="Imagem", titulo_mascara="Máscara"):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(np.squeeze(np.array(imagem)), cmap='gray')
    axes[0].set_title(titulo_imagem)
    axes[1].imshow(np.squeeze(np.array(mascara)), cmap='gray')
    axes[1].set_title(titulo_mascara)
    plt.show()

# Gerar a máscara com base na detecção de bordas
def gerar_mascara_como_canny(imagem):
    img_np = np.array(imagem)
    mascara_canny = cv2.Canny(img_np, 100, 200)
    mascara_pil = Image.fromarray(mascara_canny)
    return mascara_pil

imagem_path = '/content/3 sem mascara.png'

imagem = Image.open(imagem_path).convert("L")

mascara = gerar_mascara_como_canny(imagem)

mostrar_imagem(imagem, mascara)
```

```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms

def mostrar_imagem(imagem, mascara, titulo_imagem="Imagem", titulo_mascara="Máscara"):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(np.squeeze(np.array(imagem)), cmap='gray')
    axes[0].set_title(titulo_imagem)
    axes[1].imshow(np.squeeze(np.array(mascara)), cmap='gray')
    axes[1].set_title(titulo_mascara)
    plt.show()

# Gerar a máscara com base na detecção de bordas
def gerar_mascara_como_canny(imagem):
    img_np = np.array(imagem)
    mascara_canny = cv2.Canny(img_np, 100, 200)
    mascara_pil = Image.fromarray(mascara_canny)

    return mascara_pil

imagem_path = '/content/3 com mascara.png'

imagem = Image.open(imagem_path).convert("L")

mascara = gerar_mascara_como_canny(imagem)

# Comparação
mostrar_imagem(imagem, mascara)
```
